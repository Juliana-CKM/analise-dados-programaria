```python
tamanho_amostra = len(salarios) #verificar tamanho da amostra e atribuir variavel
tamanho_amostra
```

```python
nivel_confianca = 0.95 #atribuir nível de confiança
```

*importar biblioteca de funções estatísticas*

```python
from scipy import stats #importando biblioteca
```

```python
erro_padrao = stats.sem(salarios) #calcular erro padrão
erro_padrao
```

```python
intervalo_confianca = stats.t.interval(nivel_confianca, tamanho_amostra-1, loc=media_amostral, scale=erro_padrao)
intervalo_confianca
```

```python
from scipy.stats import chi2_contingency #chi2_contingency testa a independencia das variáveis na tabela
```

```python
def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0]     # 0 pois queremos somente o primeiro resultado
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape)-1              # padrão da fórmula
  cramer = np.sqrt(chi2/(soma*mini))
  return cramer
```

```python
cramer_coeficiente(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO']) #Aplicando Cramer. Resultado é baixo, pouca relação (por diferentes fatores)
```
