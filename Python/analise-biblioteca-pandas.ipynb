```python
def preencher_nivel (gestor,nivel):
  if gestor == 1:
    return 'Pessoa Gestora'
  else:
    return nivel
```

```python
dados3['NOVO_NIVEL'] = dados3.apply(lambda x: preencher_nivel(x['GESTOR?'], x['NIVEL']), axis=1)
```
